```python
import librosa
import matplotlib.pyplot as plt
import sounddevice as sd
```

Librosa helps with
*   Usage of audio files and loading audio recordings
*   Visualization of audio signals
*   Feature Extraction
*   Effects Creation - noise removal, Segmentation, etc.
*   Sampling



```python
# Function takes in file input and plots the signal.
def load_and_plot_signal(filename):
    y, sr = librosa.load(filename)      # Loads the Audio file. y - ndarray,sr - Sampling rate
    plt.figure(figsize=(10, 5))
    plt.plot(librosa.times_like(y), y)  # Constructs a plot of the audiofile
    plt.title('Speech Signal')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()
    return y, sr
```

```python
# Function to observe the length y - the array of samples, and the magnitude range of the signal
def observe_signal_length_magnitude(y):

    print(f"\nLength of the signal: {len(y)} samples")
    print(f"Magnitude range of the signal: {min(y)} to {max(y)}\n")
```

```python
# Function to play the Audio Segment
def play_segment(segment, sr):
    sd.play(segment, samplerate=sr)
    sd.wait()
```

```python
# Function takes in array and plots the signal.
def plot_signal(y):
    print()
    plt.figure(figsize=(10, 5))
    plt.plot(librosa.times_like(y), y)         # Constructs a plot of the audiofile
    plt.title('Speech Signal')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.show()
    return y
```

```python
def main():
    # A1. Load the recorded speech file and plot the signal
    filename = r"E:\ASEB\3rd Year\6th\Speech Processing\AI-in-Speech-Processing\Lab_1.mp3"
    y, sr = load_and_plot_signal(filename)

    # A2. Observe the length & magnitude range of the signal
    observe_signal_length_magnitude(y)

    # A3. Take a small segment of the signal and play it
    start_time = 1  # Start time in seconds
    end_time = 3    # End time in seconds
    segment = y[int(start_time * sr):int(end_time * sr)]
    play_segment(segment, sr)
    plot_signal(segment)


    # A4. Play around with different segments
    start_time_2 = 2
    end_time_2 = 8
    segment_2 = y[int(start_time_2 * sr):int(end_time_2 * sr)]
    play_segment(segment_2, sr)
    plot_signal(segment_2)
```

```python
if __name__ == "__main__":
    main()
```
